```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
labels = data['price']
conv_dates = [1 if values == 2014 else 0 for values in data.date ]
data['date'] = conv_dates
train1 = data.drop(['id', 'price'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Fix: Replace 'ls' with 'squared_error'
clf = GradientBoostingRegressor(loss='squared_error', n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Train the model
clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```

```python
import numpy as np

# Define hyperparameters
params = {
    'n_estimators': 100,  # Number of trees
    'learning_rate': 0.1,
    'max_depth': 3,
    'random_state': 42
}

# Now t_sc can be initialized correctly
t_sc = np.zeros((params['n_estimators']), dtype=np.float64)
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error

for i, y_pred in enumerate(clf.staged_predict(x_test)):
    t_sc[i] = mean_squared_error(y_test, y_pred)  # Use mean squared error instead of clf.loss_
```

```python
testsc = np.arange((params['n_estimators']))+1
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(testsc, clf.train_score_, 'b-', label='Set dev train')
plt.plot(testsc, t_sc, 'r-', label='Set dev test')
```
